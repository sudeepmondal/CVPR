## Plotting graphs with Matplotlib


```python
# Importing necessary libraries
import matplotlib.pyplot as plt

# Enable inline plotting (for Jupyter Notebook)
%matplotlib inline
```

```python
# Define x and y values
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]

# Print values for reference
print(x)
print(y)

# Set labels and title for the graph
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

# Plot the graph with default settings
plt.plot(x, y)
plt.show()
```

```python
# Plot with different colors and line width
plt.plot(x, y, color='green')  # Green line
plt.show()

plt.plot(x, y, color='red', linewidth=7)  # Red line with thickness 7
plt.show()

plt.plot(x, y, color='green', linewidth=5, linestyle='dotted')  # Dotted green line
plt.show()

plt.plot(x, y, 'b+--')  # Blue dashed line with '+' markers
plt.show()

plt.plot(x, y, color='blue', marker='+', linestyle='')  # Only markers, no line
plt.show()

plt.plot(x, y, color='blue', marker='+', linestyle='', markersize=20)  # Bigger markers
plt.show()

plt.plot(x, y, 'g<', alpha=0.5)  # Green triangle markers with transparency
plt.show()
```

```python
import numpy as np

# Create x values from -10 to 10 with a step of 0.1
x = np.arange(-10, 10, 0.1)
y1 = np.sin(x)  # Sine function values
y2 = np.cos(x)  # Cosine function values

# Plot both sine and cosine
plt.figure()
plt.plot(x, y1)
plt.plot(x, y2)
plt.legend(['Sine', 'Cosine'])  # Add legend
plt.show()
```

```python
# Another way to plot both functions together
plt.figure()
plt.plot(x, y1, x, y2)  # Plot both in one line
plt.legend(['Sine', 'Cosine'])
plt.show()
```

```python
# Creating a figure with specific size
plt.figure(figsize=(20, 5))

# First subplot (Sine Wave)
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first plot
plt.plot(x, y1)
plt.title('Sine')

# Second subplot (Cosine Wave)
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second plot
plt.plot(x, y2)
plt.title('Cosine')

plt.show()
```

```python
from IPython import display

# Create empty lists for dynamic plotting
sin = []
X = []

plt.figure()
for i in range(-50, 50, 1):  # Loop from -50 to 50
    X.append(i)  # Append x value
    sin.append(np.sin(i))  # Append sine value
    
    display.clear_output(wait=True)  # Clear previous output for animation effect
    plt.plot(X, sin)  # Plot dynamically
    plt.show()
```

```python
# Create a new figure
fig = plt.figure(figsize=(20, 5))

# First subplot (Sine Wave)
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title('Sine')

# Second subplot (Cosine Wave)
plt.subplot(1, 2, 2)
plt.plot(x, y2)
plt.title('Cosine')

# Save the figure as a PDF file
plt.savefig('test.pdf', transparent=True)
plt.show()
```
